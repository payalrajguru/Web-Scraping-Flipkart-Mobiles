```python
url = []

for i in range(1,200):
    url.append("https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=TRENDING&suggestionId=mobiles&requestId=3b1db335-1377-4654-bb77-e5c2c1f6cb9c&page="+str(i))


mobile_names = []
mobile_brands = []
ratings = []
ram = []
screen_size = []
rom = []
prices = []
front_cameras = []
back_cameras = []
colours = []
batteries_in_mAh = []
processor= []


for i in url:
    
    page=requests.get(i)
    page.text
    soup=BeautifulSoup(page.text)
    
    mobile_name = soup.find_all("div",class_="_4rR01T")
    #mobile_names = []


    for i in mobile_name:
        mobile_names.append(i.text.split('(')[0].split(',')[0].strip(' '))



    mobile_brand = soup.find_all("div",class_="_4rR01T")
    #mobile_brands = []


    for i in mobile_brand:
        mobile_brands.append(i.text.split(' ')[0].strip(','))



    rating = soup.find_all("div",class_="_3LWZlK")
    #ratings = []
    count = 0


    for i in rating:
        count=count+1
        if count<25 :
            ratings.append(i.text)




    ram_memory = soup.find_all("li",class_="rgWa7D")
    #ram = []
    

    for i in ram_memory:
        a = re.findall("(\d+)\sGB\sRAM",str(i))
        if a:
            ram.append(a[0])
            


    #screen_size = []  
    
    for i in ram_memory:
        b = re.findall("(\d+.\d+)\sinch",str(i))
        if b:
            screen_size.append(b[0])



    rom_memory = soup.find_all("li",class_="rgWa7D")
    #rom = []

    for i in rom_memory:
        a = re.findall("(\d+)\sGB\sROM",str(i))
        if a:
            rom.append(a[0])


    price = soup.find_all("div",class_="_30jeq3 _1_WHN1")
    
    #prices = []

    for i in price:
        prices.append(i.text)


    front_camera = soup.find_all('li',class_="rgWa7D")
    #front_cameras = []

    for i in front_camera:
        a = re.findall('(\d)MP Front Camera',str(i))
        if a:
            front_cameras.append(i.text.split('|')[-1].strip(' '))


    back_camera = soup.find_all('li',class_="rgWa7D")
    #back_cameras = []

    for i in back_camera:
        a = re.findall('(\d)MP Front Camera',str(i))
        if a:
            back_cameras.append(i.text.split('|')[0].strip(' '))
            
            


    color = soup.find_all('div' , class_ = '_4rR01T')
    #colours = []

    for i in color:
        colours.append(i.text.split('(')[-1].split(',')[0])
        
        



    battery = soup.find_all('li',class_="rgWa7D")
    #batteries_in_mAh = []

    for i in battery:
        a = re.findall('(\d+)\smAh',str(i))
        if a:
            batteries_in_mAh.append(i.text.split(' ')[0])



    processors = soup.find_all('li',class_="rgWa7D")
    #processor = []

    for i in processors:
        a = re.findall('[A-Za-z0-9]* Processor',str(i))
        if a:
            processor.append(i.text)
    


flip = {'mobile_name':mobile_names,
                     'brand':mobile_brands,'rating':ratings,
                     'ram':ram,'rom':rom,'prices':prices,
                     'colours':colours,'front_camera_in_MP':front_cameras,
                     'back_camera_in_MP':back_cameras,'screen_size':screen_size,
                     'batteries_in_mAh':batteries_in_mAh,'processor':processor}


flip = pd.DataFrame.from_dict(flip, orient='index')
flip= flip.transpose()
```
